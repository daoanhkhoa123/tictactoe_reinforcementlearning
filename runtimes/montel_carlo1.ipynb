```python
from src.common import RandomController, HumanController, CMDInterface, EmptyInterface, build_2clients, ClientArgs
from src.game.interface import BaseInterface
from src.game.game import Game

from src.rl_learning.montel_carlo.montel_carlo import MontelCarloClient, MontelCarloController, MontelHyperParams, build_montelclient
from src.rl_learning.montel_carlo.environment import Environment

client1, client2 = build_2clients(
    blu=ClientArgs(
        name="random bot",
        controller=RandomController(0),
        interface=BaseInterface(),
    ),
    red=ClientArgs(
        name="montel-carlo",
        controller=RandomController(),
        interface=BaseInterface(),
    ),
)

params = MontelHyperParams(
    lr=0.05,
    decay_gamma=0.95,
    exp_rate=0.2,
)

client1 =  build_montelclient(client1, params)
```

```python
print(client1.name, client1.mark_type)
print(client2.name, client2.mark_type)
```

```python
client1.controller.state_value
```

```python
from tqdm import tqdm
game = Game(client1, client2) # type: ignore

count = {}
for _ in tqdm(range(16)):
    game.reset_table()
    winner = game.hajime()
    game.owari(winner)

    count[winner] = count.get(winner, 0) + 1
    
```
